```python
import cv2
import numpy 
```

```python
def put_text_on_video(video_path):
    
    font                   = cv2.FONT_HERSHEY_DUPLEX
    bottomLeftCornerOfText = (10,500)
    fontScale              = 2
    fontColor              = (255,255,255)
    lineType               = 3

    frame_number = 900
    cap = cv2.VideoCapture(video_path)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number-1)
    frame_count = 0
    disp_range = 0
    disp_countdown = 30
    hide_countdown = 10
    
    h, w = int(cap.get(4)), int(cap.get(3))
    
    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'MJPG')
    out = cv2.VideoWriter('output.mp4',fourcc, 30.0, (w, h))
    
    while True:
        print('frame: ', frame_count)
        
        ret, frame = cap.read()
        
        if not ret:
            break
        h, w = frame.shape[:2]
        bottomLeftCornerOfText = (w-180, 90)
        if frame_count < 250:
            pass
        elif frame_count > 250 and frame_count < 1800:
            if disp_countdown > 0:
                # display 
                cv2.rectangle(frame, (w-200, 120), (w-20, 20), (0, 0, 255), cv2.FILLED)
                cv2.putText(frame,'FIRE!', bottomLeftCornerOfText, font, fontScale,fontColor,lineType)
                disp_countdown -= 1

            elif (disp_countdown == 0) and (hide_countdown > 0):
                hide_countdown -= 1

            elif disp_countdown == 0 and hide_countdown == 0:
                disp_countdown = 30
                hide_countdown = 10
        
        
        cv2.imshow('frame', frame)
        out.write(frame)
        frame_count += 1
        k = cv2.waitKey(1)
        if k == 27:
            cap.release()
            cv2.destroyAllWindows()
            out.release()
            break
```
